```python
import pandas as pd
df=pd.read_csv('weatherAUS.csv')
df
```

```python
#Data Exploration
print(df.shape,df.info())
```

```python
#Binary conversion of output
df['RainToday']=df['RainToday'].replace({'Yes':1,'No':0})
df['RainTomorrow']=df['RainTomorrow'].replace({'Yes':1,'No':0})
df.head()
```

```python
#Checking the imbalance in data
df.RainTomorrow.value_counts()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,5))
df.RainTomorrow.value_counts().plot(kind='bar',color=['skyblue','navy'])
plt.title('RainTomorrow indicator')
```

```python
#Handling the imbalance data
from sklearn.utils import resample
yes=df[df['RainTomorrow']==1]
no=df[df['RainTomorrow']==0]
oversample_yes=resample(yes,replace=True,n_samples=len(no),random_state=42)
oversample=pd.concat([no,oversample_yes])
fig = plt.figure(figsize = (8,5))
oversample.RainTomorrow.value_counts().plot(kind='bar',color=['skyblue','navy'])
plt.title('RainTomorrow indicator')
```

```python
oversample.isnull().sum()
```

```python
oversample.shape
```

```python
#imputation and transformation
oversample.select_dtypes(include=['object']).columns
```

```python
# Impute categorical var with Mode
oversample['Date'] = oversample['Date'].fillna(oversample['Date'].mode()[0])
oversample['Location'] = oversample['Location'].fillna(oversample['Location'].mode()[0])
oversample['WindGustDir'] = oversample['WindGustDir'].fillna(oversample['WindGustDir'].mode()[0])
oversample['WindDir9am'] = oversample['WindDir9am'].fillna(oversample['WindDir9am'].mode()[0])
oversample['WindDir3pm'] = oversample['WindDir3pm'].fillna(oversample['WindDir3pm'].mode()[0])
```

```python
#Converting categorical var into continuous var with labelencoder
from sklearn.preprocessing import LabelEncoder
lencoder={}
for col in oversample.select_dtypes(include=['object']).columns:
    lencoder[col]= LabelEncoder()
    oversample[col]=lencoder[col].fit_transform(oversample[col])
```

```python
oversample
```

```python
import warnings
warnings.filterwarnings("ignore")

# Multiple Imputation by Chained Equations (MICE)
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Copy the oversample data but exclude the target variable for imputation
MiceImpute = oversample.copy(deep=True) 

# Separate the features (X) and target (y)
X = MiceImpute.drop('RainTomorrow', axis=1) 
y = MiceImpute['RainTomorrow']

# Apply MICE only on the features (X), not the target
mice_imputer = IterativeImputer()
X.iloc[:, :] = mice_imputer.fit_transform(X)

# Combine the imputed features back with the target variable
X['RainTomorrow'] = y
MiceImputed=X.copy(deep=True)

```

```python
# Detecting outliers with IQR
Q1 = MiceImputed.quantile(0.25)
Q3 = MiceImputed.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Removing outliers from the dataset
MiceImputed = MiceImputed[~((MiceImputed < (Q1 - 1.5 * IQR)) |(MiceImputed > (Q3 + 1.5 * IQR))).any(axis=1)]
MiceImputed.shape
```

```python
features = MiceImpute.drop('RainTomorrow',axis=1)
target = MiceImputed['RainTomorrow']

# Split into test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25, random_state=42)

# Normalize Features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import seaborn as sns
```

```python
import time
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
def run_model(model, X_train, y_train, X_test, y_test):
    t0=time.time()
    
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
     
   
    time_taken = time.time()-t0
    print("Accuracy = {}".format(accuracy))
    
    print("Time taken = {}".format(time_taken))
    print(classification_report(y_test,y_pred,digits=5))
    
    cm = confusion_matrix(y_test, y_pred)
    # Create a heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
    plt.title('Confusion Matrix Heatmap')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    
    return model, accuracy, time_taken
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

params_lr = {'penalty': 'l2', 'solver':'liblinear'}

model_lr = LogisticRegression(**params_lr)
model_lr, accuracy_lr,tt_lr = run_model(model_lr, X_train, y_train, X_test, y_test)
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier

params_dt = {'max_depth': 16,
             'max_features': "sqrt"}

model_dt = DecisionTreeClassifier(**params_dt)
model_dt, accuracy_dt, tt_dt = run_model(model_dt, X_train, y_train, X_test, y_test)
```

```python
# Neural Network
from sklearn.neural_network import MLPClassifier

params_nn = {'hidden_layer_sizes': (30,30,30),
             'activation': 'logistic',
             'solver': 'lbfgs',
             'max_iter': 500}

model_nn = MLPClassifier(**params_nn)
model_nn, accuracy_nn,  tt_nn = run_model(model_nn, X_train, y_train, X_test, y_test)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

params_rf = {'max_depth': 16,
             'n_estimators': 150,
             'random_state': 12}

model_rf = RandomForestClassifier(**params_rf)
model_rf, accuracy_rf, tt_rf = run_model(model_rf, X_train, y_train, X_test, y_test)
```

```python
# XGBoost
import xgboost as xgb
params_xgb ={'n_estimators': 500,
            'max_depth': 16}

model_xgb = xgb.XGBClassifier(**params_xgb)
model_xgb, accuracy_xgb,  tt_xgb = run_model(model_xgb, X_train, y_train, X_test, y_test)
```

```python
# Rainfall prediction model comparison
accuracy_scores = [accuracy_lr, accuracy_dt, accuracy_nn, accuracy_rf, accuracy_xgb]

tt = [tt_lr, tt_dt, tt_nn, tt_rf, tt_xgb]

model_data = {'Model': ['Logistic Regression','Decision Tree','Neural Network','Random Forest', 'XGBoost'],
              'Accuracy': accuracy_scores,
              'Time taken': tt}
data = pd.DataFrame(model_data)
data
```
